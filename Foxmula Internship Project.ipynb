```python
while (True):

#graph displayed according to user input
    print('Available graphs:\n1.Movies released per year(MY)\n2.Movies released per genre per year(MGY)')
    print('3.Budget spent per year(BY)\n4.Prominent genre of the year(PGY)')
    print('5.Highest rated movie of the year(HMY)\n6.Lowest rated movie of the year(LMY)')
    print('7.End code execution(END)\n')

    s=input('Choose one of the above (Type abbreviation given in bracket) : ')
    import matplotlib.pyplot as plt

    if s=='MY':
        m7=[] ; m8=[]; m9=[] ; m10=[]; m11=[]
        for i in range(len(yr)):
            if yr[i]==2007:
                m7.append(m[i])
            elif yr[i]==2008:
                m8.append(m[i])
            elif yr[i]==2009:
                m9.append(m[i])
            elif yr[i]==2010:
                m10.append(m[i])
            elif yr[i]==2011:
                m11.append(m[i])
        print('\n_MOVIES RELEASED PER YEAR_')      
        x1=[2007,2008,2009,2010,2011]
        y1=[len(m7),len(m8),len(m9),len(m10),len(m11)]
        plt.title('Number of movies released during the year')
        plt.bar(x1,y1,width= 0.5, color=['red','blue'])
        plt.xlabel('Year'); plt.ylabel('No. of movies released')
        plt.show()

    elif s=='MGY':
        print('\n_MOVIES RELEASED PER GENRE PER YEAR_')
        print('\nMovies released per genre : 2007')
        c=0 ; ad=0; ac=0; ho=0; dr=0; ro=0;t=0
        for i in range(len(yr)):
            if yr[i]==2007:
                if g[i]=='Action':
                    ac+=1
                elif g[i]=='Adventure':
                    ad+=1
                elif g[i]=='Comedy':
                    c+=1
                elif g[i]=='Horror':
                    ho+=1
                elif g[i]=='Drama':
                    dr+=1
                elif g[i]=='Romance':
                    ro+=1
                elif g[i]=='Thriller':
                    t+=1
        x2=['Action','Adventure','Comedy','Horror','Thriller','Drama','Romance']
        y2=[ac,ad,c,ho,t,dr,ro] ; colors=['r','blue','y','orange','grey','green','pink']
        fig, ax=plt.subplots()
        plt.pie(y2,colors=colors,labels=x2,startangle=90, radius=2,autopct='%1.1f%%')
        plt.show()
        print('\nMovies released per genre : 2008')
        c=0 ; ad=0; ac=0; ho=0; dr=0; ro=0;t=0
        for i in range(len(yr)):
            if yr[i]==2008:
                if g[i]=='Action':
                    ac+=1
                elif g[i]=='Adventure':
                    ad+=1
                elif g[i]=='Comedy':
                    c+=1
                elif g[i]=='Horror':
                    ho+=1
                elif g[i]=='Drama':
                    dr+=1
                elif g[i]=='Romance':
                    ro+=1
                elif g[i]=='Thriller':
                    t+=1
        y2=[ac,ad,c,ho,dr,ro,t]
        plt.pie(y2,colors=colors,labels=x2,startangle=90, radius=2,autopct='%1.1f%%')
        plt.show()
    
        print('\nMovies released per genre : 2009')
        c=0 ; ad=0; ac=0; ho=0; dr=0; ro=0;t=0
        for i in range(len(yr)):
            if yr[i]==2009:
                if g[i]=='Action':
                    ac+=1
                elif g[i]=='Adventure':
                    ad+=1
                elif g[i]=='Comedy':
                    c+=1
                elif g[i]=='Horror':
                    ho+=1
                elif g[i]=='Drama':
                    dr+=1
                elif g[i]=='Romance':
                    ro+=1
                elif g[i]=='Thriller':
                    t+=1
        y2=[ac,ad,c,ho,dr,ro,t]
        plt.pie(y2,colors=colors,labels=x2,startangle=90, radius=2,autopct='%1.1f%%')
        plt.show()  
        print('\nMovies released per genre : 2010')      
        c=0 ; ad=0; ac=0; ho=0; dr=0; ro=0;t=0
        for i in range(len(yr)):
            if yr[i]==2010:
                if g[i]=='Action':
                    ac+=1
                elif g[i]=='Adventure':
                    ad+=1
                elif g[i]=='Comedy':
                    c+=1
                elif g[i]=='Horror':
                    ho+=1
                elif g[i]=='Drama':
                    dr+=1
                elif g[i]=='Romance':
                    ro+=1
                elif g[i]=='Thriller':
                    t+=1
        y2=[ac,ad,c,ho,dr,ro,t]
        plt.pie(y2,colors=colors,labels=x2,startangle=90, radius=2,autopct='%1.1f%%')
        plt.show()
        print('\nMovies released per genre : 2011')
        c=0 ; ad=0; ac=0; ho=0; dr=0; ro=0;t=0
        for i in range(len(yr)):
            if yr[i]==2011:
                if g[i]=='Action':
                    ac+=1
                elif g[i]=='Adventure':
                    ad+=1
                elif g[i]=='Comedy':
                    c+=1
                elif g[i]=='Horror':
                    ho+=1
                elif g[i]=='Drama':
                    dr+=1
                elif g[i]=='Romance':
                    ro+=1
                elif g[i]=='Thriller':
                    t+=1
        y2=[ac,ad,c,ho,dr,ro,t]
        plt.pie(y2,colors=colors,labels=x2,startangle=90, radius=2,autopct='%1.1f%%')
        plt.show()
    
    elif s=='BY':

        print('\n_BUDGET SPENT PER YEAR_')
        b7=0;b8=0;b9=0;b10=0;b11=0
        x1=[2007,2008,2009,2010,2011]
        for i in range(len(yr)):
            if yr[i]==2007:
                b7=b7+b[i]
            elif yr[i]==2008:
                b8=b8+b[i]
            elif yr[i]==2009:
                b9=b9+b[i]
            elif yr[i]==2010:
                b10=b10+b[i]
            elif yr[i]==2011:
                b11=b11+b[i]
        y2=[b7,b8,b9,b10,b11]
        plt.title('Movie budget per year')
        plt.xlabel('Year') ; plt.ylabel('Million $')
        plt.plot(x1,y2,color='orange',linewidth=2,marker='o',markerfacecolor='blue',markersize=10)
        plt.show()
    elif s=='PGY':

        print('\n_PROMINENT GENRE OF THE YEAR_')
        import statistics as st
        G7=[] ; G8=[]; G9=[] ; G10=[]; G11=[] 
        x1=[2007,2008,2009,2010,2011]
        for i in range(len(yr)):
            if yr[i]==2007:
                G7.append(g[i])
            elif yr[i]==2008:
                G8.append(g[i])
            elif yr[i]==2009:
                G9.append(g[i])            
            elif yr[i]==2010:
                G10.append(g[i])               
            elif yr[i]==2011:
                G11.append(g[i])
        f=[G7.count(st.mode(G7)),G8.count(st.mode(G8)),G9.count(st.mode(G9)),G10.count(st.mode(G10)),G11.count(st.mode(G11))]
        gh=[st.mode(G7),st.mode(G8),st.mode(G9),st.mode(G10),st.mode(G11)]
        plt.title('Prominent genre of the year')
        plt.xlabel('Year') ; plt.ylabel('No. of movies released')
        colour=['red','green','blue','orange','black']
        plt.bar(x1,f,color=colour,label=st.mode(G7))
        plt.bar(x1,f,color=colour,label=st.mode(G8))
        plt.bar(x1,f,color=colour,label=st.mode(G9))
        plt.bar(x1,f,color=colour,label=st.mode(G10))
        plt.bar(x1,f,color=colour,label=st.mode(G11))
        for i in range(len(f)):
            plt.text(x = x1[i]-0.25 , y = f[i], s = gh[i], size = 12)
    
        plt.subplots_adjust(bottom= 0.2, top = 0.98)
        plt.show()

    elif s=='HMY':
        print('\n_HIGHEST RATED MOVIE OF THE YEAR_')
        x1=[2007,2008,2009,2010,2011]
        G7H=[] ; G8H=[]; G9H=[] ; G10H=[]; G11H=[]
        gh7=[];gh8=[];gh9=[];gh10=[];gh11=[]
        for i in range(len(yr)):
            if rt[i]>85 and ra[i]>85:
                if yr[i]==2007:
                    G7H.append((rt[i]+ra[i])/2);gh7.append(m[i])
                elif yr[i]==2008:
                    G8H.append((rt[i]+ra[i])/2);gh8.append(m[i])
                elif yr[i]==2009:
                    G9H.append((rt[i]+ra[i])/2);gh9.append(m[i])            
                elif yr[i]==2010:
                    G10H.append((rt[i]+ra[i])/2) ;gh10.append(m[i])              
                elif yr[i]==2011:
                    G11H.append((rt[i]+ra[i])/2);gh11.append(m[i])
            
        final=G7H+G8H+G9H+G10H+G11H
        final_m=gh7+gh8+gh9+gh10+gh11
        finalH=[max(G7H),max(G8H),max(G9H),max(G10H),max(G11H)]
        finalm=[final_m[final.index(max(G7H))],final_m[final.index(max(G8H))],final_m[final.index(max(G9H))],final_m[final.index(max(G10H))],final_m[final.index(max(G11H))]]
        plt.title('Highest rated movie of the year')
        plt.xlabel('Year') ; plt.ylabel('Rating')
        plt.plot(x1,finalH,color='green',marker='o',markerfacecolor='blue',markersize=8)

        for i in range(len(finalH)):
            plt.text(x = x1[i]-0.25 , y = finalH[i]+0.1, s = (finalm[i],finalH[i]), size = 12)
        plt.subplots_adjust(bottom= 0.2, top = 0.98)
        plt.show()
        
    elif s=='LMY':
        G7L=[] ; G8L=[]; G9L=[] ; G10L=[]; G11L=[]
        ml7=[];ml8=[];ml9=[];ml10=[];ml11=[]
        for i in range(len(yr)):
            if rt[i]<85 and ra[i]<85:
                if yr[i]==2007:
                    G7L.append((rt[i]+ra[i])/2);ml7.append(m[i])
                elif yr[i]==2008:
                    G8L.append((rt[i]+ra[i])/2);ml8.append(m[i])
                elif yr[i]==2009:
                    G9L.append((rt[i]+ra[i])/2);ml9.append(m[i])            
                elif yr[i]==2010:
                    G10L.append((rt[i]+ra[i])/2) ;ml10.append(m[i])              
                elif yr[i]==2011:
                    G11L.append((rt[i]+ra[i])/2);ml11.append(m[i])
                    
        print('\n_LOWEST RATED MOVIE OF THE YEAR_')
        x1=[2007,2008,2009,2010,2011]
        final2=G7L+G8L+G9L+G10L+G11L
        final_m2=ml7+ml8+ml9+ml10+ml11
        finalL=[min(G7L),min(G8L),min(G9L),min(G10L),min(G11L)]
        finalm2=[final_m2[final2.index(min(G7L))],final_m2[final2.index(min(G8L))],final_m2[final2.index(min(G9L))],final_m2[final2.index(min(G10L))],final_m2[final2.index(min(G11L))]]
        plt.title('Lowest rated movie of the year')
        plt.xlabel('Year') ; plt.ylabel('Rating')
        plt.plot(x1,finalL,color='red',marker='o',markerfacecolor='blue',markersize=8)

        for i in range(len(finalL)):
    
            if i==4:
                plt.text(x = x1[i]-0.25 , y = finalL[i]+1.5, s = (finalm2[i],finalL[i]), size = 12)
            else:
                plt.text(x = x1[i]-0.25 , y = finalL[i]+0.1, s = (finalm2[i],finalL[i]), size = 12)
        
        plt.subplots_adjust(bottom= 0.2, top = 0.98)
        plt.show()
    
    #to terminate the program (infinite loop)
    elif(s=='END'):
        print('\n_____Program Terminated_______')
        break
        
    #MESSAGE DISPLAYED TO USER IF INPUT DOESN'T MATCH THE OPTIONS
    else:
        print('\nINVALID INPUT!!!!  ENTER AVAILABLE ABBREVIATIONS ONLY!!!  \n\n')
```
